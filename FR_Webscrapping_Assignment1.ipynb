# BOOK PAGE 

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/")
page
```

```python
page.content
```

```python

soup = BeautifulSoup(page.content)
soup
```

```python

print(soup.prettify())
```

```python
book = soup.find_all('h4',class_="italic")
book
```

```python
book_name = [] 
for i in book:
    book_name.append(i.text.replace('\n',''))
book_name 
```

```python
len(book_name)
```

```python
author = soup.find_all('p',class_="sans bold")
author
```

```python
author_name = [] 
for i in author:
    author_name.append(i.text.replace('\n',''))
author_name
```

```python
len(author_name)
```

```python
genre = soup.find_all('p',class_="genre-links hidden-phone")
genre
```

```python
genre_name = [] 
for i in genre:
    genre_name.append(i.text.replace('\n',''))
genre_name
```

```python
len(genre_name)
```

```python
review= soup.find_all('p',class_="excerpt")
review
```

```python
review_name = [] 
for i in review:
    review_name.append(i.text.replace('\n',''))
review_name
```

```python
len(review_name)
```

```python
import pandas as pd
booksite = pd.DataFrame({})
booksite['book'] = book_name
booksite['author'] = author_name
booksite['genre'] = genre_name
booksite['review'] = review_name
booksite.head()
```
